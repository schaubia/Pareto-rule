Pareto principle, also referred to as the “80/20 rule”, states that approximately 80% of the consequences are attributable to around 20% of the causative factors. Dr. Joseph M. Juran, a management consultant and also an engineer, named the principle after the Italian economist Vilfredo Pareto. Pareto made several observations reflecting the disproportion of causes and effects in physical life. Back in 1896, Pareto published a paper after observing that roughly 20% of the pea plants in his garden produced 80% of the peas. Later on, he also revealed that 80% of the property in Italy was owned by 20% of the citizens and that 80% of the income in a country was earned by 20% of the population.

Juran generalised Pareto’s findings authoring the phrase “The vital few and the trivial many”. As an engineer he noticed that a small number of defects caused the bulk of issues and began to apply the Pareto principle to quality management issues. The Pareto principle is a special case of the wider phenomenon of Pareto distributions. It can be noted that:<br>
* 80% of revenue comes from 20% of customers.
* 20% of employees produce 80% of a company’s results.
* 20% of customers produce 80% of the revenue.
* 80% of results come from 20% of effort and time spent.

Asserting that often the majority of results comes from a minority of inputs, the Pareto principle can be a useful tool to help prioritizing actions and allocating resources efficiently. It can be applied in management, economics and business, but also in computer and natural science, to improve productivity, make better decisions, and yield more accurate assessment. A visual representation of the 80/20 rule is the Pareto chart. It lists the factors in a given situation in a descending order of their frequency, highlighting those having the greatest impact. This provides essential and easily interpretable insight on where the focus should be put on.


### Pareto chart
Let's say participants in an online math course spend a total of around 1000 hours studying maths from online resources. We can suppose approximately 80% of their time is dedicated to understanding mathematical analysis.

```python
# create a list of random numbers that sum up to approximately 1000 and one of them holds 80% of thier 
values = [np.random.randint(780, 820)]
nums = list(np.random.randint(0, 200, size = 4))
for num in nums:
    values.append(int(num/sum(nums)*200))
```

```python
values
```

```python
#create a dataframe with the list of values as hours per each math field
df = pd.DataFrame({"hours": values})
df.index = ["analysis", "arithmetics", "algebra", "geometry", "combinatorics"]
df.sort_values("hours", ascending = False, inplace = True)
```

```python
df
```

```python
#calculate the cumulative percentage of the hours spent on studying for each math branch
cum_percent = np.cumsum(df.hours)

#create the Pareto chart
#create subplots - one for the bar chart with math categories and one for the line plot with cumulative percents
fig, ax = plt.subplots(figsize=(12,6))

sns.barplot(df.index, df.hours)

ax2 = ax.twinx()
ax2.plot(df.index, cum_percent, color = "C2", marker = "D")

#set the labels of the y axis to show the cumulative percentage
percents = ax2.get_yticks()
ax2.set_yticklabels(["{}%".format(0.1* i) for i in percents], color = "C2")

plt.title("Pareto chart for hours spent on each math branch")
plt.savefig("Pareto chart")
plt.show()
```
